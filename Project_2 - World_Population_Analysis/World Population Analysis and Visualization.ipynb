# 3. Exploratory Data Analysis

```python
# Statistical Summary of numerical columns in the dataframe

pd.options.display.float_format = '{:.3f}'.format
describe = df.describe()
describe
```

```python
# Total population in 2022

print('Total population in 2022: ', df['2022 Population'].sum())
```

```python
# Record of the country with the most population in year 2022

max_population = df[df['2022 Population'] == df['2022 Population'].max()]
max_population
```

```python
# Record of the country with the highest area

highest_area = df[df['Area (km²)'] == df['Area (km²)'].max()]
highest_area
```

```python
# World population numerical breakdown by continent

pd.options.display.float_format = '{:.3f}'.format
continent = df.groupby('Continent')
continent.mean(numeric_only = True)
```

```python
# Top 20 most populated countries in 2022

highest_twenty = df[df['Rank'] <= 20].sort_values(by = 'Rank')
highest_twenty
```

```python
# Population growth rate for each country from 2020 to 2022

df['Population Growth Rate'] = (df['2022 Population'] - df['2020 Population']) / df['2020 Population']
df[['Country/Territory', 'Population Growth Rate']]
```

```python
# Correlation of each column

df.corr(numeric_only = True)
```

# 4. Data Visualization

```python
# Top 10 most populated countries in 2022 by continent

africa = df[df.Continent == 'Africa']
afr = africa.sort_values(by = '2022 Population', ascending = True).tail(10)
asia = df[df.Continent == 'Asia']
asi = asia.sort_values(by = '2022 Population', ascending = True).tail(10)
europe = df[df.Continent == 'Europe']
eur = europe.sort_values(by = '2022 Population', ascending = True).tail(10)
north_america = df[df.Continent == 'North America']
nor_am = north_america.sort_values(by = '2022 Population', ascending = True).tail(10)
oceania = df[df.Continent == 'Oceania']
oce = oceania.sort_values(by = '2022 Population', ascending = True).tail(10)
south_america = df[df.Continent == 'South America']
sou_am = south_america.sort_values(by = '2022 Population', ascending = True).tail(10)

fig , ax = plt.subplots(nrows=2,ncols=3, figsize = (12,9))
plt.subplots_adjust(wspace = 1.7)

plt.suptitle('Top 10 Most Populated Countries in Each Continent')
ax[0,0].barh(afr['Country/Territory'], afr['2022 Population'] / 10**6)
ax[0,0].set_title('Africa')
ax[0,1].barh(asi['Country/Territory'], asi['2022 Population'] / 10**6)
ax[0,1].set_title('Asia')
ax[0,2].barh(eur['Country/Territory'], eur['2022 Population'] / 10**6)
ax[0,2].set_title('Europe')
ax[1,0].barh(nor_am['Country/Territory'], nor_am['2022 Population'] / 10**6)
ax[1,0].set_title('North America')
ax[1,1].barh(sou_am['Country/Territory'], sou_am['2022 Population'] / 10**6)
ax[1,1].set_title('South America')
ax[1,2].barh(oce['Country/Territory'], oce['2022 Population'] / 10**6)
ax[1,2].set_title('Oceania')
plt.show()
```

```python
# World population percentage by continent

df_cont = df[['Continent','World Population Percentage']].groupby(['Continent']).sum(numeric_only=True).sort_values(by = 'World Population Percentage', ascending=False)

# Graphing using pie chart
labels = df_cont.index.tolist()
colors = ['aquamarine', 'violet', 'orange', 'orangered', 'lightblue', 'yellow']

plt.figure(figsize=(10, 7))
plt.pie(df_cont['World Population Percentage'], labels=labels, autopct='%1.1f%%', colors = colors)
plt.legend()
plt.title('World Population Percentage by Continent')
plt.show()
```

```python
# Percentage of countries for each country

area = df[['Continent','Area (km²)']].groupby(['Continent']).sum(numeric_only=True).sort_values(by = 'Area (km²)', ascending=False)

# Graphing:
labels = area.index.tolist()
colors = ['aquamarine', 'violet', 'orange', 'orangered', 'lightblue', 'yellow']

plt.figure(figsize=(10, 7))
plt.pie(area['Area (km²)'], labels=labels, autopct='%1.1f%%', colors = colors)
plt.legend()
plt.title('Area in Square Kilometers by Continent')
plt.show()
```

```python
# Trend of world population trend (1970-2022)

total_pop = df[['1970 Population', '1980 Population', '1990 Population',
                '2000 Population', '2010 Population', '2015 Population',
                '2020 Population', '2022 Population']].sum()

# Plotting the population trend

plt.figure(figsize=(15, 6))
plt.plot(total_pop, marker = 'o')
plt.xlabel('Population Year')
plt.ylabel('Count')
plt.title('World Population Trend (in Billion)')
plt.show()
```
